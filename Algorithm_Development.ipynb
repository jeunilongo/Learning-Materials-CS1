##What is an **Algorithm**?

An algorithm is a step-by-step procedure or set of instructions designed to solve a specific problem or accomplish a task. It outlines a clear sequence of actions to achieve a particular goal, ensuring that the process is logical and can be repeated consistently.

### Example: Finding the Maximum Number in a List

1. **Understanding the problem**
* Input: A list of numbers (e.g., [4, 7, 1, 9, 3]
* Output: The largest number (e.g., 9)

2. **Designing with Pseudocode**
* Pseudocode is a plain-language representation of the algorithm. It focuses on logic without worrying about syntax.


```
START
1. Initialize a variable 'max_number' to the first number in the list.
2. For each number in the list:
    a. If the number is greaater than 'max_number', update 'max_number'.
3. Return 'max_number' as the result.
END
```



````python
#Python Implementation

def find_max(numbers):
  # Step 1: Initialize max_number
  max_number = numbers[0]

  # Step 2: Iterate through th list
  for num in numbers:
    if num > max_number:
        max_number = num # Update max_number if a larger number is found

  return max_number

numbers = [4, 7, 1, 9, 3]
print(find_max(numbers))  # Output: 9
````

##Python Syntax and Structure

Python is a high-level, easy-to-read programming language with a clean and straightforward syntax.

**Key Characteristics include:**
1. Indentation for blocks: Python  uses indentation (whitespace) to define blocks of code instead of braces ({}) or other delimiters.


```
if x>0:
  print("Positive number") # Indented code belongs to the if block
```



2. Comments: Use the # symbol for single-line comments.

```
# This is a comment
print("Hello, World!") # Prints a message
```



3. Case Sensitivity: Variable and function names are case-sensitive (MyVariable and myvariable are different)

4. Dynamic Typing: Variables don't require explicit type declarations.


```
x = 10 # Integer
y = "Hi" # String
```



5. Common Structures: Python uses familiar constructs like variables, operators, loops, conditionals, and functions.

## How Algorithms are Represented

1. **Functions**: Functions encapsulate algorithms to make code reusable and modular

````python
def find_max(a,b):
  return a if a>b else b

print(find_max(5,10))
````

2. **Loops**: Loops allow repetitive tasks to be handled efficiently.

````python
for i in range(5): # Iterates 5 times
  print(i)  # Prints number 0 to 4
````

3. **Conditions**: Conditional statements (if, elif, else) handle decision-making.

````python
age = 20
if age < 18:
  print("Minor")
else:
  print("Adult")
````

4. **Combining Structures**: Algorithms often combine these elements to solve problems effectively. For example, finding the sum of even numbers in a list.

````python
numbers = [1,2,3,4,5]
total = 0
for num in numbers:
  if num % 2 == 0: # Check if the number is even or not
    total += num

print(total)  # Output: 6
````
